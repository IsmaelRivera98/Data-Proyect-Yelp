### Eliminando los duplicados de la columna user_id

Esto lo realizamos debido a que tenemos:

valores unicos = 1987897; registros totales = 2105597

Realizar la eliminacion de los registros duplicados, representaria el 5%, lo cual podemos concluir que no representa una cantidad considerable del data set original.

```python
# Usar el método 'duplicated' para identificar registros duplicados en 'id'
duplicados = data[columna].duplicated(keep=False)

# Filtrar el DataFrame para mantener solo los registros no duplicados
registros_unicos = data[~duplicados]
```

```python
registros_unicos = registros_unicos.reset_index(drop=True)
```


### Analizando la columna review_count

```python
# Encontrar el valor máximo en la columna 'review_count'
max_review_count = registros_unicos['review_count'].max()

print("El valor máximo en la columna 'review_count' es:", max_review_count)
```

```python
# Contar los registros que tienen valores iguales a cero en 'review_count'
count_zeros = (registros_unicos['review_count'] == 0).sum()

print(f"La cantidad de registros con valores iguales a cero en 'review_count' es: {count_zeros}")
```

#### Realizamos la eliminación de los registros con review_count =0. Esto se realiza debido a que no tiene interaccion con la pltaforma.

```python
# Eliminar los registros con valor igual a cero en 'review_count'
registros_unicos = registros_unicos[registros_unicos['review_count'] != 0]
```

#### Normalizamos la columna review_count en otra columna de nombre review_count_normalized para posteiormente visualizar

```python
# Calcular el mínimo y el máximo de la columna 'review_count'
min_review_count = registros_unicos['review_count'].min()
max_review_count = registros_unicos['review_count'].max()

# Aplicar la normalización Min-Max a la columna 'review_count'
registros_unicos['review_count_normalized'] = (registros_unicos['review_count'] - min_review_count) / (max_review_count - min_review_count)
```
